### Logit from Mastering Machine Learning

```python
# Make a prediction with coefficients
def predict(row, coefficients):
	yhat = coefficients[0]
	for i in range(len(row)-1):
		yhat += coefficients[i + 1] * row[i]
#	print 1.0 / (1.0 + exp(-yhat))        
	return 1.0 / (1.0 + exp(-yhat))

# Estimate logistic regression coefficients using stochastic gradient descent
def coefficients_sgd(train, l_rate, n_epoch):
	coef = [0.0 for i in range(len(train[0]))]
	cost_history = []
	for epoch in range(n_epoch):
		sum_error = 0
		for row in train:
			yhat = predict(row, coef)
			error = row[-1] - yhat
			sum_error += error**2
			coef[0] = coef[0] + l_rate * error * yhat * (1.0 - yhat)
			for i in range(len(row)-1):
				coef[i + 1] = coef[i + 1] + l_rate * error * yhat * (1.0 - yhat) * row[i]
		#print('>epoch=%d, lrate=%.3f, error=%.3f' % (epoch, l_rate, sum_error))
		cost_history.append(sum_error)        
	return coef,cost_history

# Calculate coefficients
# we create a dataset with umpy putting X features along with the  Y target values...
dataset = np.concatenate((X,y.reshape((y.shape[0],1))),axis=1)
l_rate = 0.1
n_epoch = 1000
coef,cHistory = coefficients_sgd(dataset, l_rate, n_epoch)
print(coef)

```

```python
def calAccuracy(target,y_h):
    size = len(target)
    correct = 0
    for i in range(size):
        if target[i] == y_h[i]:
            correct += 1
    return (correct/float(size))*100.0
W = coef#np.array([-3.0387985111396993, 8.8951547598883192, 2.790680310948916])
y_h = []
for row in dataset:
    y_h.append(round(predict(row,W)))
#np.concatenate((y_h,y), axis=1)
y_h = np.array(y_h,dtype=int)
print y[:10]
print y_h[:10]
print "Accuracy:",calAccuracy(y,y_h),"%"

c1 = X[y_h==0]
c2 = X[y_h==1]
label = np.hstack((y_h[y_h==0],y_h[y_h==1]))
data = np.vstack((c1,c2))
plt.scatter(data[:,0],data[:,1],c=label)
#plt.plot(X,z)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()

plt.plot(cHistory)
plt.show()
```
