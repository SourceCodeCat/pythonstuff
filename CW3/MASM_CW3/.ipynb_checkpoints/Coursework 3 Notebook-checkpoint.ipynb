**Question 1.1 [10 points]** 

What percentage of the Tweets from this dataset fall within the 50 states (including Alaska, Hawaii, and DC) of the US?

````python
import pandas as pd
import geopandas as gpd
%matplotlib inline  
import fiona
import shapely
from shapely.geometry import Point
# read-in the Tweets CSV from where we will get the Longitude and LAtitude of the Tweet
tweets = pd.read_csv("data/tweets.csv")
# read in the shape file using  Geopandas, and creating a GeoDataFrame
# the sahpe contains the polygons for the 51 states ( 51 if DC is considered as state)
US = gpd.read_file("data/states_21basic/states.shp")
# extract the  Longitude and Latitude from the tweets so we can create geometry points
tweets_points=tweets[["Longitude","Latitude","RTs"]].values
# create geometry points that can be used to see if they intersect any stste polygon
tweets_points = [{"RTs":t[2], "geometry": Point(t[0],t[1])} for t in tweets_points]
tweets_gdp = gpd.GeoDataFrame(tweets_points,crs={'init' :'epsg:4269'})
# check  from which states is each tweet querying it in the US shapes
states_with_tweets = gpd.sjoin(tweets_gdp, US, how="inner", op='intersects')
# save the number of tweets in the US
numberOfStatesWithTweets = float(len(states_with_tweets))
# save the number of tweets in total
numberOfTweets = float(len(tweets_points))
# printing the percent  of tweets inside the US from the tweets CSV
print "If we take only those tweets that matched a US Polygon then:"
print "%.2f %% of tweets are in the US" % ((numberOfStatesWithTweets/numberOfTweets)*100.0)
print "If we calculate the percentage with the tweets filtered only by the Country Column of the CSV file:"
print "%.2f %% of tweets(filtered by Country) are in the US" % ((len(tweets.Country[tweets.Country=="US"])/numberOfTweets)*100.0)
````

**Question 1.2 [10 points]** 

For those Tweets that are within the US boundaries, create a *cloropleth* map of the ***United States*** showing the number of Tweets per State (treat DC as a state in this case).  Include a legend showing the color scale for the number of Tweets.

````python
# Now from the dataset I try to group the tweets and count them so we can know how many tweets are for each STATE_NAME
tweet_count = states_with_tweets[["STATE_NAME","index_right"]].groupby("index_right").count().rename(columns={"STATE_NAME":"num_tweets"})
# merge the counts to the base dataframe, so we can plot it
US = US.merge(tweet_count, left_index=True, right_index=True, how="left")
# merging the data  will produce some NaN, because there are only 49 states with tweets, 
# those NaN will be replaced by zeros
US = US.fillna(0)
````

````python
# Plotting the Geopandas Dataset.. shoosing the tweet count to create a chloropleth
US.plot(figsize=(30,20),legend=True, column="num_tweets", cmap="Reds",scheme='fisher_jenks')
````

**Question 1.3 [10 points]** 

Create a *cloropleth* map of the ***United States*** showing the total sum of retweets for the Tweets per state (and DC).  Include a legend showing the color scale for the number of Retweets.

````python
# we proceed the same with the Tweet's re-tweets, but doing a sum of the re-tweets per state, merge them and fill any NaN
# produced by the merge.
tweet_rts = states_with_tweets[["RTs","index_right"]].groupby("index_right").sum().rename(columns={"RTs":"num_RTs"})
US = US.merge(tweet_rts, left_index=True, right_index=True, how="left")
US = US.fillna(0)
````

````python
# plot the  re-tweet numbers as a chloropeth
US.plot(figsize=(30,20),legend=True, column="num_RTs", cmap="OrRd", scheme='fisher_jenks')
````

**Question 1.4 [10 points]** 

Find the top 3 states with the *highest density* of Tweets (from the dataset) per area.  

````python
#from the dataset we can  group by state the number of tweets and set the name of the count as num_tweets
tweets_per_state = states_with_tweets[["STATE_NAME","index_right"]].groupby("STATE_NAME").count().rename(columns={"index_right":"num_tweets"})
#Printing the top 3 states with high density of tweets:
tweets_per_state.sort_values(by="num_tweets",ascending=False)[0:3]
````

## Part 2 [NLP]

This data contains the text of the tweet (in the 'tweet content' column of the tweets.csv provided). In this section we will be focusing on this column.

**Question 2.1 [10 points]** 

Load the 'Tweet content' column into NLTK. 

Make sure that when you read in the .csv you use the following syntax:


```all=pd.read_csv('tweets.csv', encoding='iso-8859-1')```

This encoding will ensure that the text is read in clean of strange symbols.

Show your code and document it, ensuring that you have produced an NLTK.Text.text object. 

````python
import pandas as pd
import nltk
tweets = pd.read_csv('data/tweets.csv', encoding='iso-8859-1')
# converting the Tweet_content column to a string, and removing the index numbers...
tcontent = tweets["Tweet content"].to_string(index=False)
# Creating a Text requires the string to be tokenized..
tcontent_token = nltk.word_tokenize(tcontent)
# creating a Text var.
tweetsText = nltk.Text(tcontent_token)
print "type of tweetsText Variable is : %s" % type(tweetsText)
````

** Question 2.2 [10 points] **

Use the object in Question 2.1 to calculate the 50 most frequent words.

````python
# creating a var to store the frequence distribution of the tokens in the text
tweetsText_fdist = nltk.FreqDist(tweetsText)
print "50 Most Frequent Words:"
# to print the 50 most frequent, we can make use of the method  most_common(top number qwe want)
# from the results obtained we can  create a dataframe :) to show the results
pd.DataFrame(tweetsText_fdist.most_common(50),columns=["word","frequency"])
````

** Question 2.3 [5 points]**

Plot a cumulative frequency distribution of the most frequent words.

````python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
# we can use the frequency distribution optained earlier and plot it right away...passing as apramter we want
# the top 50 in a cumulative way.
tweetsText_fdist.plot(50, cumulative=True)
````

## Part 3 [Graphs]


In this section we are going to focus on the connections between different hashtags, and treat them as a graph in order to visualise the relationships between hashtags that occur together.

** Question 3.1 [5 points]**

The file coachella_hashtags.csv contains a sample of the tweets that have the #coachella in them. This is a small sample of tweets with the #coachella and which have at least one more # in them. These hashtags have been extracted to separate columns: tag1, tag2, tag3.

Read this data in and using tag1 and tag2 load this data into a networkX Graph object and draw the graph. Use apropriate options and colors.

````python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
# reading in the CSV file using pandas
hashtags = pd.read_csv('data/coachella_hashtags.csv', encoding='iso-8859-1')
#creating a graph using the  dataframe and choosing the data we want to use (tag1, tag2), also this will be a
# simple graph
hashtags_graph = nx.from_pandas_dataframe(hashtags,'tag1','tag2',create_using=nx.Graph())
# we can set the name of the graph
hashtags_graph.name="Coachella HashTags"
plt.figure(figsize=(10,8))
# plotting the graph, a smaller font_size seems to  fit in the words better
nx.draw(hashtags_graph, with_labels=True, node_size=150, node_color="white", font_size=8)
````

** Question 3.2 [10 points] **

Using the graph created in 3.1 provide the following information about the graph:

- Numeber of nodes and number of edges
- The nodes connected to ```#lcdcoundsystem``` (***updated***)
- The node with the highest clustering measure
- The node with the highes closeness centrality measure
- Find the ***cliques*** of the network (***updated***)

````python
import operator
#print "Number of Nodes: %s\nNumber of Edges: %s" % (hashtags_graph.number_of_nodes(), hashtags_graph.number_of_edges())
print nx.info(hashtags_graph)
# the  method neighbors returns all the nodes connected to #lcdsoundsystem
print "Nodes connected to #lcdcoundsystem : %s" % ",".join(hashtags_graph.neighbors("#lcdcoundsystem"))
# the clustering method will do the job here
cluster_dict = nx.clustering(hashtags_graph)
print "The node with the highest clustering measure : %s with a value of: %s" % max(cluster_dict.iteritems(), key=operator.itemgetter(1))
# for this we will use the closeness_centrality method
closeness_dict = nx.closeness_centrality(hashtags_graph)
print "The node with the highes closeness centrality measure : %s with a value of: %s" % max(closeness_dict.iteritems(), key=operator.itemgetter(1))
print "The cliques of the network:"
pd.DataFrame(list(nx.find_cliques(hashtags_graph)),columns=["Hashtag 1","Hashtag 2","Hashtag 3"])
````

** Question 3.4 [10 points] **

The coachella_hashtags.csv file contained a small sample of tweets, which are not representative of all the #coachella tweets.

Go back to tweets.csv and filter the tweets to only those tweets that have the hashtag #coachella in the 'tweet content'. 

Create your own larger version of coachella_hashtag.csv using all the hashtags in the tweet text and on apply it to all the tweets that include #coachella already in their tweet text. 

Use this data to populate a new Graph object using tag1 and tag2 that you have created.

1. Provide the code you have used, make sure there are sufficient explanations.
2. Report back the number of nodes and edges.

````python
# from tweets.csv get "Tweet content" that has a "#coachella" in the text
withHashTag = (tweets["Tweet content"][tweets["Tweet content"].str.contains("#coachella ")])
# once we have all the tweets, lets get rid of text and leave only the  hashtags 
withHashTag = [[t for t in row.split() if t.startswith("#")] for row in withHashTag]
# we create a DataFrame with the tags, 
withHashTag_df = pd.DataFrame(withHashTag)
# as we can obtain alot of columns( each tweet can have many hashtags) we calculate the column names
columns=["tag%s" % (i+1) for i in range(len(withHashTag_df.columns))]
# assign the column names to the data frame
withHashTag_df.columns=columns
#  get rid of those rows that doesn't have a value in both coumns (tag1 and tag2)
withHashTag_df = withHashTag_df[withHashTag_df.tag2.notnull()]
# create an nx Graph from the dataframe by grabbing the first two columns only (there are tweets with >11  tags!!!)
withHashtags_graph = nx.from_pandas_dataframe(withHashTag_df,"tag1","tag2",create_using=nx.Graph())
#print the edges and nodes info
print nx.info(withHashtags_graph)
````

````python
plt.figure(figsize=(10,8))
nx.draw(withHashtags_graph, with_labels=True, node_size=150, node_color="white", font_size=10)
````
